# matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x/10,y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = "red", linestyle = '-.') # :  -.  --  -
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color = "red", linestyle = '-.') # :  -.  --  -
plt.xlim(-2.5,12.5)
plt.ylim(0,1.5)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy) # :  -.  --  -
y1 = np.full(shape = (100,1), fill_value=1.03)
y2 = np.full(shape = (100,1), fill_value=-1.03)
plt.plot(x,y1,color = "black", linestyle = '--')
plt.plot(x,y2,color = "black", linestyle = '--')
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x,siny, label="sin(x)")
plt.plot(x,cosy, label = "cos(x)") # :  -.  --  -
y1 = np.full(shape = (100,1), fill_value=0.53)
y2 = np.full(shape = (100,1), fill_value=-1.03)
plt.plot(x,y1,color = "black", linestyle = '--')
plt.plot(x,y2,color = "black", linestyle = '--')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title("test(1)", color = 'black')
plt.legend()
```
